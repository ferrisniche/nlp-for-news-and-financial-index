```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import statsmodels.api as sm
```

```python
file_path = 'C:/Users/zz/Desktop/time series12.15/time series project/'
os.chdir(file_path)
data = pd.read_csv('topic proportion time series.csv')
index = pd.read_csv('Index.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
index['Date'] = pd.to_datetime(index['Date'])
index = index.set_index('Date')
```

```python
ree=index[['returns']]
data=pd.merge(data,ree,on='Date', how='inner')
data=data.sort_index()
```

```python
topic4p=np.percentile(data[['topic4']].values, 0)
data4 = data[data.topic4>topic4p]
```

```python
X = data4[['topic4']].values
X = sm.add_constant(X)
y = data4[['returns']]
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
print(model.summary2())
```

```python
plt.scatter(data4[['topic4']].values,data4[['returns']])
plt.plot(data4[['topic4']].values, model.fittedvalues, color='r',label='fitted line')
plt.axhline(0, color='green', linestyle='--', linewidth=1)
plt.xlabel('Finance Topic')
plt.ylabel('Stock Return')
plt.legend()
plt.show()
```

```python
residuals = model.resid

# 绘制残差的自相关图
fig, ax = plt.subplots(figsize=(8, 6))
sm.graphics.tsa.plot_acf(residuals, lags=20, ax=ax)
plt.title("Autocorrelation of Residuals")
plt.show()
```
